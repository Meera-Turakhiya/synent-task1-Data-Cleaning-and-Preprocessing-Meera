**Problem Statement**

The dataset contains missing values, inconsistent data types, and unnecessary information that can affect analysis and machine learning performance. The objective of this task is to clean and preprocess the dataset to improve its quality and make it suitable for further analysis.

```python
import pandas as pd
```

```python
df=pd.read_csv("Titanic_Dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df.dtypes
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df['Survived']=df['Survived'].astype('category')
```

```python
df.rename(columns={
    'PassengerId': 'Passenger_ID',
    'Pclass': 'Passenger_Class',
    'Sex': 'Gender'
}, inplace=True)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv("Cleaned_Titanic_Dataset.csv",index=False)
```

**Key Insights**

Missing values are common in real-world datasets and require proper handling.

Mean imputation is useful for numerical features such as Age.

Mode imputation is suitable for categorical features such as Embarked.

Removing highly incomplete columns improves data quality.

Clean and structured data provides a strong foundation for machine learning and analysis.
